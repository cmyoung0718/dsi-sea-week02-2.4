```python
# Value count of songs on chart by genre for first week

pd.value_counts(df['genre'])
```

```python
# Total count of songs on chart for 1st week
sum(pd.value_counts(df.ix[:,'x1st.week']))
```

```python
# Count of rock songs on the chart for 1st week
sum_num = sum(pd.value_counts(df.ix[df['genre'] == 'Rock']['x1st.week']))
# Mean place on chart for rock for the first week
mn = np.mean(df.ix[df['genre'] == 'Rock']['x1st.week'])
print mn
# Average place for a rock song on the chart for week one
print 'The average place on the Billboard Chart for a rock track in week 1 was',mn

# might as well do it for all in a function
gens = []
for i in df['genre']:
    if i not in gens:
        gens.append(i)
weeks = [s for s in df.columns.values if "week" in s]        

def avg_place(gen,week):
    mn = np.mean(df.ix[df['genre'] == gen][week])
    print 'The average place on the Billboard Chart for a',gen, 'track in', week, 'was',mn

for i in gens:
    for j in weeks:
        avg_place(i,j)
```
